# Monte Carlo Simulation

In this step, you will run Monte Carlo Simulations for your portfolio to model portfolio performance at different retirement ages. 

Complete the following steps:
1. Calculate the daily returns for the SPY and AGG closing prices.
2. Calculate volatility for both the SPY and AGG closing prices.
3. Find the last day's closing price for both stocks and save those as variables.
4. Run a Monte Carlo Simulation of at least 500 iterations and generate at least 30 years of closing prices

### HINTS:
There are 252 trading days per year, so the number of records to generate for each Monte Carlo run will be 252 days * 30 years

```python
# Calculate the daily roi for the stocks
# Drop extra columns 
data.drop(columns=['open', 'high', 'low', 'volume'], level=1, inplace=True)
data.index = data.index.date
data
```

```python
daily_returns = data.pct_change()
daily_returns.head()
```

```python
# Calculate volatility
# Calculate Mean
avg_daily_return_agg = daily_returns.mean()['AGG']['close']
avg_daily_return_spy = daily_returns.mean()['SPY']['close']
print(avg_daily_return_agg)
print(avg_daily_return_spy)
```

```python
#Calculate std dev
std_dev_daily_return_agg = daily_returns.std()['AGG']['close']
std_dev_daily_return_spy = daily_returns.std()['SPY']['close']
print(std_dev_daily_return_agg)
print(std_dev_daily_return_spy)
```

```python
# Save the last day's closing price
# YOUR CODE HERE
agg_last_price=data['AGG']['close'][-1]
spy_last_price=data['SPY']['close'][-1]
```

```python
# Setup the Monte Carlo Parameters
number_simulations = 500
number_records = 252 * 30
monte_carlo = pd.DataFrame()
portfolio_cumulative_returns = pd.DataFrame()
```

```python
# Run the Monte Carlo Simulation
for x in range(number_simulations):
    simulated_agg_prices = [agg_last_price]
    simulated_spy_prices = [spy_last_price]
    for i in range(number_records):
            simulated_agg_price = simulated_agg_prices[-1] * (1 + np.random.normal(avg_daily_return_agg, std_dev_daily_return_agg))
            simulated_spy_price = simulated_spy_prices[-1] * (1 + np.random.normal(avg_daily_return_spy, std_dev_daily_return_spy))
            simulated_agg_prices.append(simulated_agg_price)
            simulated_spy_prices.append(simulated_spy_price)
    monte_carlo["AGG prices"] = pd.Series(simulated_agg_prices)
    monte_carlo["SPY prices"] = pd.Series(simulated_spy_prices)
    simulated_daily_returns = monte_carlo.pct_change()
    #print(simulated_daily_returns)
    # Set the portfolio weights (60% SPY; 40% AGG)
    weights = [0.6, 0.4]
    # Use the `dot` function with the weights to multiply weights with each column's simulated daily returns
    portfolio_daily_returns = simulated_daily_returns.dot(weights)
    # Calculate the normalized, cumulative return series
    portfolio_cumulative_returns[x] = (1 + portfolio_daily_returns.fillna(0)).cumprod()
    # Print records from the DataFrame
portfolio_cumulative_returns.head()
```

```python
# Visualize the Simulation
plot_title = f"Simulations of Cumulative Portfolio Return Trajectories Over the Next 30 Years"
portfolio_cumulative_returns.plot(legend=None, title=plot_title)
```

```python
# Select the last row for the cumulative returns (cumulative returns at 30 years)
ending_cumulative_returns = portfolio_cumulative_returns.iloc[-1, :]
ending_cumulative_returns.head()
```

```python
# Select the last row for the cumulative returns (cumulative returns at 20 years)
ending_cumulative_return_20yrs = portfolio_cumulative_returns.iloc[-2520, :]
ending_cumulative_return_20yrs.head()
```

```python
# Display the 90% confidence interval for the ending returns
confidence_interval = ending_cumulative_returns.quantile(q=[0.05, 0.95])
confidence_interval
```

```python
# Visualize the distribution of the ending returns
plt.figure();
ending_cumulative_returns.plot(kind='hist', density=True, bins=10)
plt.axvline(confidence_interval.iloc[0], color='r')
plt.axvline(confidence_interval.iloc[1], color='r')
```
